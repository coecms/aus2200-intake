# Demo for the AUS2200 intake catalogue

This catalogue is a work in progress and we welcome feedback on how to bets structure it to facilatet queries.

```python
import intake
```

```python
cat = intake.open_catalog('/g/data/ua8/Working/packages/aus2200-intake/catalogue.yaml')
```

List all the datasets available.

```python
list(cat)
```

At the moment we have one dataset for each separate simulation and a combined dataset which includes all simulations.

## Example of dataset for a single simulation: canberra03

```python
canberra03 = cat['canberra03']
```

```python
canberra03.description
```

```python
canberra03.df.head()
```

To see all available variables

```python
canberra03.unique()['variable']
```

Retrieve `tas`

```python
sub = canberra03.search(variable='tas')
```

Create a dictionary of datasets

```python
ds_dict = sub.to_dataset_dict()
```

```python
ds_dict.keys()
```

Load 1 hourly tas

```python
tas_ds = ds_dict['v1-0.1hr.tas']
tas_ds
```

## Example with combined dataset

```python
comb = cat['combined']
```

```python
sub = comb.search(variable='tas', frequency='1hr')
```

```python
ds_dict = sub.to_dataset_dict()
```

```python
ds_dict.keys()
```

```python
blacksat_tas = ds_dict['blacksat.v1-0.1hr.tas']
blacksat_tas
```
